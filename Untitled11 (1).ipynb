```python
# Convert to datetime FIRST
df_taxi['tpep_pickup_datetime'] = pd.to_datetime(df_taxi['tpep_pickup_datetime'])
df_taxi['tpep_dropoff_datetime'] = pd.to_datetime(df_taxi['tpep_dropoff_datetime'])

# Calculate duration (must use BOTH datetimes)
df_taxi['trip_duration'] = (
    df_taxi['tpep_dropoff_datetime'] - df_taxi['tpep_pickup_datetime']
).dt.total_seconds() / 60  # minutes

# Extract day/hour features (before dropping!)
df_taxi['pickup_day'] = df_taxi['tpep_pickup_datetime'].dt.day
df_taxi['pickup_hour'] = df_taxi['tpep_pickup_datetime'].dt.hour

# NOW drop the raw timestamps (prevent leakage)
df_taxi.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime'], axis=1, inplace=True)
```

```python

# 3. Define robust outlier bounds (using IQR)
def treat_outliers(df, column):
    Q1 = df[column].quantile(0.05)  # Using 5th/95th percentiles for robustness
    Q3 = df[column].quantile(0.95)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# 4. Apply to all numeric columns
numeric_cols = ['trip_distance', 'trip_duration', 'extra',
               'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge']

for col in numeric_cols:
    print(f"\nTreating outliers in {col}:")
    print(f"Original shape: {df_taxi.shape}")
    df_taxi = treat_outliers(df_taxi, col)
    print(f"New shape: {df_taxi.shape}")

# 5. Final sanity checks
print("\nFinal dataset shape:", df_taxi.shape)
print("\nOutlier-treated stats:")
print(df_taxi[numeric_cols].describe(percentiles=[0.01, 0.25, 0.5, 0.75, 0.99]))
```

```python
# 1. Remove physically impossible values
df_taxi = df_taxi[
    (df_taxi['trip_distance'] > 0.05) &     # Minimum 0.05 miles
    (df_taxi['trip_duration'] > 0.5) &      # Minimum 30 seconds
    (df_taxi['trip_duration'] < 180)        # Maximum 3 hours
]

# 2. Drop constant-value columns (no predictive power)
df_taxi.drop(['mta_tax', 'improvement_surcharge', 'tolls_amount'], axis=1, inplace=True)

# 3. Robust outlier capping (preserves more data than removal)
def cap_outliers(df, col):
    q1 = df[col].quantile(0.01)
    q99 = df[col].quantile(0.99)
    df[col] = np.where(df[col] < q1, q1, df[col])
    df[col] = np.where(df[col] > q99, q99, df[col])
    return df

for col in ['trip_distance', 'trip_duration', 'extra', 'tip_amount']:
    df_taxi = cap_outliers(df_taxi, col)

# 4. Final verification
print("\nFinal Statistics:")
print(df_taxi.describe(percentiles=[0.01, 0.25, 0.5, 0.75, 0.99]))
print("\nMissing Values Check:")
print(df_taxi.isnull().sum())
```

```python
# Cap extreme (but plausible) weather values
climate_df['wspd'] = climate_df['wspd'].clip(upper=30)  # Max 30 m/s wind
climate_df['tmax'] = climate_df['tmax'].clip(upper=25)   # Max 25C in Jan
```

reat question! The wind direction (wdir) preprocessing is crucial because it's a circular variable (where 0 = 360), and raw degrees can mislead models. Here's why we engineer it:

1. Problem with Raw Degrees (0-360)
Mathematical Issue: 359 and 1 are just 2 apart, but their numeric difference (358) suggests they're opposites.

Model Misinterpretation: Linear models (and even some DNNs) treat 359 as "larger" than 1, which is physically incorrect.

```python
# 1. First create the cardinal directions
climate_df['wind_dir'] = pd.cut(
    climate_df['wdir'],
    bins=[0, 90, 180, 270, 360],
    labels=['N', 'E', 'S', 'W'],
    right=False
)

# 2. One-hot encode the wind directions
climate_df = pd.get_dummies(climate_df, columns=['wind_dir'], prefix='wind')

# 3. NOW create the engineered features
climate_df['temp_range'] = climate_df['tmax'] - climate_df['tmin']
climate_df['wind_chill'] = 13.12 + 0.6215*climate_df['tavg'] - 11.37*(climate_df['wspd']**0.16) + 0.3965*climate_df['tavg']*(climate_df['wspd']**0.16)

# 4. Select final columns
final_climate_cols = [
    'tavg', 'tmin', 'tmax',         # Temperature
    'wspd',                         # Wind speed
    'pres',                         # Pressure
    'is_cold', 'is_windy',          # Binary flags
    'temp_range', 'wind_chill',     # Engineered features
    'wind_N', 'wind_E', 'wind_S', 'wind_W'  # One-hot encoded
]

climate_df = climate_df[final_climate_cols]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def create_mlp_model():
    model = Sequential([
        # Input layer + 1st hidden layer (64 neurons)
        Dense(64, activation='relu', input_shape=(24,)),

        # 2nd hidden layer (32 neurons)
        Dense(32, activation='relu'),

        # Output layer (1 neuron for regression)
        Dense(1)
    ])

    model.compile(
        optimizer='adam',
        loss='mse',
        metrics=['mae']
    )
    return model

# Initialize model
mlp_model = create_mlp_model()
mlp_model.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

def create_dnn_model():
    model = Sequential([
        # Input layer + 1st hidden layer (128 neurons)
        Dense(128, activation='relu', input_shape=(24,)),
        Dropout(0.2),  # Randomly disable 20% of neurons to prevent overfitting

        # 2nd hidden layer (64 neurons)
        Dense(64, activation='relu'),
        Dropout(0.2),

        # 3rd hidden layer (32 neurons)
        Dense(32, activation='relu'),

        # Output layer
        Dense(1)
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )
    return model

# Initialize
dnn_model = create_dnn_model()
dnn_model.summary()
```
